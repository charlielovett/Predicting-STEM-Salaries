```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
data = pd.read_csv('./Data/salary_data_cleaned.csv')
import warnings
warnings.filterwarnings("ignore")
```

```python
data.head()
```

```python
data.totalyearlycompensation.describe()
```

```python
model = sm.ols(formula = 'totalyearlycompensation~company+title+location+yearsofexperience+yearsatcompany+gender+cityid+dmaid+Masters_Degree+Bachelors_Degree+Doctorate_Degree+Highschool+Some_College+Race_Asian+Race_White+Race_Two_Or_More+Race_Black+Race_Hispanic', data = data).fit()
model.summary()
```

```python
pred = model.predict(data)
```

```python
(np.abs(data.totalyearlycompensation - pred)).mean()
```
