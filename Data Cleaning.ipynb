```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('./Data/salary_data.csv')
import warnings
warnings.filterwarnings("ignore")
```

```python
data = data.drop(columns = ["timestamp", "basesalary", "stockgrantvalue", "otherdetails", "rowNumber", "bonus"])
data = data.loc[data.gender.notnull() & data.Race.notnull() & data.Education.notnull()]
data = data.drop(columns = ["Race", "Education"])
data.company = data.company.str.lower()
data.company = data.company.str.replace(" ", "")
data.company = data.company.str.replace("-", "")
data.company = data.company.str.replace(",", "")
data.company = data.company.str.replace(".", "")
```
